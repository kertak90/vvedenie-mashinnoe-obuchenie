```python
import math
import numpy as np
import matplotlib.pyplot as plt

```

```python
X = np.random.normal(loc=1, scale=10, size=(1000, 50))
print(X)
```

```python
m = np.mean(X, axis=0)
print(m)
```

```python
std = np.std(X, axis=0)
print(std)
X_norm = ((X - m) / std)
print(X_norm)
```
